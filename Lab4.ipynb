```python
class Vaccine:

    def __init__(self,vcname,country, minage):
        self.vcname = vcname
        self.country = country
        self.mindays = minage


class Person:
    def __init__(self,name,age,occu="General Citizen"):
        self.name = name
        self.age = age
        self.occu = occu
        self.dose = ''
        self.eligible = True
        if self.age < 25 and self.occu != 'Student':
            self.eligible = False
            

    def pushVaccine(self, other, dose = None):
        if not self.eligible:
            print(f'Sorry {self.name}, Minimum age for taking vaccines is 25 years now.')
        elif dose == None and self.eligible:
            self.dose = other.vcname
            print(f"1st dose done for {self.name}")
            self.d1 = 'Given'
            self.d2 = f'Please come after {other.mindays} days'
        else:
            if self.dose != other.vcname:
                print(f'Sorry {self.name}, you can’t take 2 different vaccines')
            else:
                print(f"2nd dose done for {self.name}")
                self.d1 = 'Given'
                self.d2 = 'Given'
    

    def showDetail(self):
        print(f'Name: {self.name} Age: {self.age} Type: {self.occu}')
        print(f'Vaccine name: {self.dose}')
        print(f'1st dose: {self.d1}')
        print(f'2st dose: {self.d2}')

#Write your code here
astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra) # 1st dose done for Bob
print("=================================")
p1.showDetail() # Name: Bob Age: 21 Type: Student {...}
print("=================================")
p1.pushVaccine(sin, "2nd Dose") # Sorry Bob, you can’t take 2 different vaccines
print("=================================")
p1.pushVaccine(astra, "2nd Dose") # 2nd dose done for Bob
print("=================================")
p1.showDetail() # Name: Bob Age: 21 Type: Student {...}
print("=================================")
p2 = Person("Carol", 23, "Actor") 
print("=================================")
p2.pushVaccine(sin) 
print("=================================")
p3 = Person("David", 34) 
print("=================================")
p3.pushVaccine(modr) # 1st dose done for David
print("=================================")
p3.showDetail() # Name: David Age: 34 Type: General Citizen {...}
print("=================================")
p3.pushVaccine(modr, "2nd Dose") # 2nd dose done for David

# Expected Output:
# =================================
# 1st dose done for Bob
# =================================
# Name: Bob Age: 21 Type: Student 
# Vaccine name: AstraZeneca
# 1st dose: Given
# 2nd dose: Please come after 60 days
# =================================
# Sorry Bob, you can’t take 2 different vaccines
# =================================
# 2nd dose done for Bob
# =================================
# Name: Bob Age: 21 Type: Student 
# Vaccine name: AstraZeneca
# 1st dose: Given
# 2nd dose: Given
# =================================
# Sorry Carol, Minimum age for taking vaccines is 25 years now.
# =================================
# 1st dose done for David
# =================================
# Name: David Age: 34 Type: General Citizen
# Vaccine name: Moderna
# 1st dose: Given
# 2nd dose: Please come after 30 days 
# =================================
# 2nd dose done for David
```
