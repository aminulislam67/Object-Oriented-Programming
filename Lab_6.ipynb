```python
class Assassin:
    c=0
    def __init__(self,name,success_rate):
        self.name=name
        self.success_rate=success_rate
        Assassin.c+=1
    @classmethod    
    def failureRate(cls,name,failure_rate):
        obj=cls(name,100-failure_rate)
        return obj
    @classmethod    
    def failurePercentage(cls,name,rate):
        sr=int(rate[:-1:])
        sr1=100-int(sr)
        obj=cls(name,sr1)
        return obj
    def printDetails(self):        
        print("Name:",self.name)
        print(f'Success rate: {self.success_rate}%')
        print("Total number of Assassin:", Assassin.c)
        
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', "10%")
akabane.printDetails()
```

```python
class Cat:
    Number_of_cats=0
    def __init__(self,color="white",position="sitting"):
        self.color=color
        self.position = position
        Cat.Number_of_cats+=1
    @classmethod
    def no_parameter(cls): 
        return cls("White","sitting")
    @classmethod
    def first_parameter(cls,clr): 
        return cls(clr)
    @classmethod
    def second_parameter(cls, position): 
        return cls("White", position)  
    def changeColor(self,color): 
        self.color = color       
    def printCat(self):
        print(f'{self.color} cat is {self.position}')
        
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
import math
class Cylinder:
    radius = 5
    height = 18
    def __init__(self, radius, height):
        print("Default radius=",Cylinder.radius,"and height=",Cylinder.height)
        self.radius=radius
        self.height=height
        Cylinder.radius=self.radius
        Cylinder.height=self.height
        print("Updated: radius=",self.radius,"and height=",self.height)
    @staticmethod
    def area(radius, height):
        area = 2*math.pi* float(radius)*float(radius)+2*math.pi* float(radius)*float(height)
        print("Area:",area)
    @staticmethod
    def volume(radius, height):
        volume =math.pi*float(radius)*float(radius)*float(height)
        print("Volume:",volume)
    @classmethod
    def swap(cls,radius,height):
        return Cylinder(height, radius)
    @classmethod
    def changeFormat(cls,string):
        radius,height =string.split('-')
        return Cylinder(float(radius),float(height))

c1 = Cylinder(0, 0)
Cylinder.area(c1.radius, c1.height)
Cylinder.volume(c1.radius, c1.height)
print("===============================")
c2 = Cylinder.swap(8, 3)
c2.area(c2.radius, c2.height)
c2.volume(c2.radius, c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius, c3.height)
c3.volume(c3.radius, c3.height)
print("===============================")
Cylinder(0.3, 5.56).area(Cylinder.radius, Cylinder.height)
print("===============================")
Cylinder(3, 5).volume(Cylinder.radius, Cylinder.height)
```

```python
class SultansDine:
    new_l=[]
    branch_name=''
    branch_sell=0   
    def __init__(self,branch_name):
        self.branch_name=branch_name
        SultansDine.new_l.append(self)
    def sellQuantity(self,quantity):
        if quantity<10:
            self.Branch_sell=quantity*300
        elif quantity<20:
            self.Branch_sell=quantity*350
        else:
            self.Branch_sell = quantity*400
    def branchInformation(self):
        print("Branch Name:",self.branch_name)
        print("Branch Sell:",self.Branch_sell,"Taka")
    def details():
        new_l=SultansDine.new_l
        sell=0
        branchNo=len(new_l)      
        for item in new_l:
            sell+=item.Branch_sell
        print("Total number of branch(s):", branchNo)
        print("Total Sell:",sell,"Taka")       
        if branchNo==0:
            return
        for branch in new_l:
            print("Branch Name:",branch.branch_name,"Branch sell:",branch.branch_sell,"Taka")
            total_sell=round(branch.Branch_sell*100.0/sell,2)
            print(f'Branch consists of total sell\'s: {total_sell}%')

SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
